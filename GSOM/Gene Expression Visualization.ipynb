```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE, Isomap
from sklearn.decomposition import PCA, KernelPCA
from AEGSOM import GSOM
```

```python
def visualizeW1(opt_W1, vis_patch_side, hid_patch_side):

    """ Add the weights as a matrix of images """

    figure, axes = plt.subplots(nrows = 1,
                                              ncols = hid_patch_side)
    index = 0

    for axis in axes.flat:

        """ Add row of weights as an image to the plot """

        image = axis.imshow(np.repeat(opt_W1[index, :],50).reshape(vis_patch_side, 50),cmap = plt.cm.gray,
                            interpolation = 'nearest')
        axis.set_frame_on(False)
        axis.set_axis_off()
        index += 1

    """ Show the obtained plot """

    plt.show()
```
