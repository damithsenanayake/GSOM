```python
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
gene_expr = pd.read_csv('~/data/gene_expr.csv')
gene_expr
```

```python
gene_expr.shape
```

```python
data = np.array(gene_expr).T
```

```python
expr = data[1:].astype(float)
```

```python
expr
```

```python
means = np.log2(expr+0.5).mean(axis = 0)
```

```python
vars_ = np.log2(expr+0.5).var(axis = 0)
```

```python
cors = np.polyfit(means, vars_, deg = 4)
fit = np.poly1d(cors)
```

```python
plt.scatter(means, vars_, marker='3', color = 'black')
plt.plot(np.sort(means), fit(np.sort(means)), c = 'red')
plt.show()
```

```python
HDGs=np.where((vars_-fit(means)) > 0.125)[0]
```

```python
gene_labels = np.array(gene_expr[gene_expr.columns[0]])[HDGs]
```

```python
gene_counts = np.array(gene_expr)[:, 1:][HDGs]
```

```python
gene_counts.shape
```

```python
gene_labels.shape
```

```python
pd.DataFrame(data=np.append(np.array([gene_labels]).T, gene_counts, axis=1), columns = gene_expr.columns)
```
