```python
def  triangulate(grid):
    adj = np.zeros((grid.shape[0], grid.shape[0]))
    tri = Del(grid)
    chull = ConvexHull(grid).simplices
    print chull
    for simplegrid in tri.simplices:
        for vert in simplegrid:
            for vert2 in simplegrid:
                adj[vert][vert2] = 1
    for line in chull:
        adj[line[0]][line[1]]=0
        adj[line[1]][line[0]]=0
    
    return adj
```

```python
def get_neighbors(point, grid):
    G = triangulate(grid)    
    return np.where(G[point])[0]
```

```python
X = np.array(pd.read_csv('/home/senanayaked/data/spiral.csv', header=None))[:, :2]#np.random.random((100, 3))
X += abs(X.min())
X /= X.max()
C = np.ones(X.shape)*X.mean()#np.random.random((X.shape[0], X.shape[1]))
grid = np.random.random((X.shape[0], 2))
```

```python
try:
    for it in range(100):
        for x in X[np.random.permutation(X.shape[0])]:
            # find BMU    
            bmu = np.argmin(np.linalg.norm(C -x, axis = 1))
            neighbors = get_neighbors(bmu, grid)
#             print neighbors
            dists = np.linalg.norm(grid[neighbors] - grid[bmu], axis = 1)
            moving_amounts = np.array([np.exp(-dists/2)]).T
    #         print moving_amounts

            C[neighbors] += (x- C[neighbors]) * np.exp(-it/100) * moving_amounts

            # move neighbors closer


            moving_directions = grid[bmu] - grid[neighbors]

            grid[neighbors] += 0.01*np.exp(-it/100)*moving_amounts * moving_directions


            not_neighbors = np.setdiff1d(np.array(range(X.shape[0])), neighbors)

            dists = np.linalg.norm(grid[not_neighbors] - grid[bmu], axis = 1)
            moving_amounts = np.array([np.exp(-dists/5)]).T
            moving_directions = grid[bmu] - grid[not_neighbors]
            grid[not_neighbors] -=0.01*np.exp(-it/100)* moving_amounts * moving_directions
except :
    print 'error'
    pass
#predict 
out = []
for x in X:
    bmu = np.argmin(np.linalg.norm(C -x, axis = 1))
    out.append(grid[bmu])
out = np.array(out)
```

```python
plt.scatter(out.T[0], out.T[1], c=plt.cm.Set1(np.array(range(X.shape[0])).astype(float)/X.shape[0]), edgecolors='none', alpha=0.5)
plt.plot(out.T[0], out.T[1], c='grey', alpha=0.5)
plt.show()
```
