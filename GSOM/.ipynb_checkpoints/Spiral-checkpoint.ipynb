```python
import numpy as np
import pandas as pd
from sgsom import GSOM
from AEGSOM import GSOM as GS
import matplotlib.pyplot as plt
from visualize import scatter_basic, scatter_gradient
from sklearn.manifold import TSNE, MDS, ISOMAP
```

```python
%pylab inline
```

```python
spiral = pd.read_csv('~/data/spiral.csv', header=None)
X = np.array(spiral).astype(float)
X += abs(np.amin(X))
X /= X.max()
```

```python
X_r = X[np.random.permutation(X.shape[0])]
```

```python
X_r
```

```python
adj = np.zeros((X_r.shape[0], X_r.shape[0]))

visited = []
for i in range(X_r.shape[0]):
    visited.append(i)
    neighbors = np.where(np.linalg.norm(X_r - X_r[i], axis = 1)<0.05)[0] 
    adjacent = np.setdiff1d(neighbors, np.array(visited))#np.array(list(neighbors) - visited)
    adj[i][adjacent]=1
   
```

```python
adj = np.zeros((X_r.shape[0], X_r.shape[0]))

visited = []

for i in range(X_r.shape[0]):
    visited.append(i)
    neighbors = np.argsort(np.linalg.norm(X_r - X_r[i], axis = 1))[:2]
    adjacent = np.setdiff1d(neighbors, np.array(visited))
    print adjacent

    adj[i][adjacent] = 1
```

```python
S = []

for i in range(X.shape[0]):
    if not adj[:, i].any():
        S.append(i)
S
```

```python
L = []
```

```python
while len(S)>0:
    n = S.pop()
    L.append(n)
    
    for m in np.where(adj[n])[0]:
        adj[n][m]=0
        if not adj[:, m].any():
            S.append(m)
```

```python
order = np.array(L)
```

```python
scatter_gradient(X.T[0], X.T[1])
```

```python
plt.scatter(X_r[order].T[0], X_r[order].T[1],c = plt.cm.Set1(np.array(range(X.shape[0])).astype(float)/X.shape[0]), edgecolors='none')
plt.plot(X_r[order].T[0], X_r[order].T[1], c = 'grey')
plt.show()
```

```python
# gsom = GSOM(dims=2, min_nodes=10, max_nodes = 1000, fd = 0.9, sf=0.8, radius=1.41)
gsom = GS(dims=2, hid = 5, X =X, fd=0.9, sf=0.999, min_nodes=10, max_nodes=100000, radius=1.41, gaussian=True, nei=True, map_init=10)
```

```python
for i in range(1):
    print '\nbatch ', (i+1)
    gsom.train_batch(X=X_r, lr=1*np.exp(-i/2), iterations=2, prune=False)

grid, hits = gsom.predict(X)

scatter_gradient(grid.T[0], grid.T[1])
```
